```python
# import the necessary packages
from imutils import paths
import requests
import cv2
import os
from datetime import datetime
```

load Mars bars images

```python
urlFile = "/home/anna/Documents/MarsBarNotMarsBar/moreMarsBarUrls.txt"
pathToImgDir = "/home/anna/Documents/MarsBarNotMarsBar/marsBar/"
total = 0
```

```python
with open(urlFile) as f:
    content = f.readlines()
f.close()
# remove whitespace characters like `\n` at the end of each line
rows = [x.strip().split("\n") for x in content] 
rowsCleaned = []
for url in rows:
    rowsCleaned.append(url[0])
rowsCleaned[0]
```

```python
numURLS = len(rowsCleaned)
print "Number of URLs obtained: {}".format(numURLS)
```

```python
%%time
# loop the URLs
i = 0
now = datetime.now()
nowStr = now.strftime("%Y-%m-%d %H:%M")
for url in rowsCleaned:
    p = url
    try:
        r = requests.get(url, timeout=60)
        p = pathToImgDir + "{}{}.jpg".format(nowStr, str(total).zfill(8))
        f = open(p, "wb")
        f.write(r.content)
        f.close()
        print("[INFO] downloaded: {}".format(p))
        total += 1
    except:
        print("[INFO] error downloading {}...skipping".format(p))
```

```python
print "sucessfully downloaded {} images from {}".format(total, numURLS)
```
